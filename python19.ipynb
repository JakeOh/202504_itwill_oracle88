## 생존자 비율

*   전체 승객 중 남성 생존자 비율, 여성 생존자 비율
    *   남성 생존자 비율 = (남성 생존자 수) / (전체 승객 수)
    *   여성 생존자 비율 = (여성 생존자 수) / (전체 승객 수)

*   생존자들 중 남성 생존자 비율, 여성 생존자 비율
    *   남성 생존자 비율 = (남성 생존자 수) / (전체 생존자 수)
    *   여성 생존자 비율 = (여성 생존자 수) / (전체 생존자 수)

*   생존자들 중 성인 생존자 비율, 어린이 생존자 비율
    *   성인 생존자 비율 = (성인 생존자 수) / (전체 생존자 수)
    *   어린이 생존자 비율 = (어린이 생존자 수) / (전체 생존자 수)

```python
# 전체 승객수
# num_of_total = titanic.shape[0]
num_of_total = len(titanic)
num_of_total
```

```python
# 남성 생존자
num_of_male_alive = len(titanic[(titanic.sex == 'male') & (titanic.alive == 'yes')])
num_of_male_alive
```

```python
# 여성 생존자
num_of_female_alive = len(titanic[(titanic.sex == 'female') & (titanic.alive == 'yes')])
num_of_female_alive
```

```python
ratio_male_alive = num_of_male_alive / num_of_total
ratio_male_alive
```

```python
ratio_female_alive = num_of_female_alive / num_of_total
ratio_female_alive
```

```python
# 전체 생존자
num_of_total_alive = len(titanic[titanic.alive == 'yes'])
num_of_total_alive
```

```python
print('남성 생존자 비율 =', num_of_male_alive / num_of_total_alive)
print('여성 생존자 비율 =', num_of_female_alive / num_of_total_alive)
```

```python
# 어린이 생존자
num_of_child_alive = len(titanic[(titanic.who == 'child') & (titanic.alive == 'yes')])
num_of_child_alive
```

```python
# 성인 생존자
num_of_adult_alive = len(titanic[(titanic.who != 'child') & (titanic.alive == 'yes')])
num_of_adult_alive
```

```python
print('성인 생존자 비율 =', num_of_adult_alive / num_of_total_alive)
print('어린이 생존자 비율 =', num_of_child_alive / num_of_total_alive)
```

## 객실 등급별 생존자 비율

*   (1등급 객실 생존자 비율) = (1등급 객실 생존자 수) / (전체 생존자 수)
*   (2등급 객실 생존자 비율) = (2등급 객실 생존자 수) / (전체 생존자 수)
*   (3등급 객실 생존자 비율) = (3등급 객실 생존자 수) / (전체 생존자 수)

*   (1등급 객실 생존자 비율) = (1등급 객실 생존자 수) / (1등급 객실 승객수)
*   (2등급 객실 생존자 비율) = (2등급 객실 생존자 수) / (2등급 객실 승객수)
*   (3등급 객실 생존자 비율) = (3등급 객실 생존자 수) / (3등급 객실 승객수)
